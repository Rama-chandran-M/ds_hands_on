```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('dataset.csv')
df.head(2)
```

## Adding Price Category Column (Low,Mid,High)

```python
def price_category(price):
    if price <= 75:
        return "Low"
    elif price <= 150:
        return "Mid"
    else:
        return "High"

df["price_category"] = df["final_price_usd"].apply(price_category)
```

## Adding Discount Status Boolean Column 

```python
df["discount_applied_status"] = (df["discount_percent"] != 0).astype(int)
```

```python
df.to_csv("dataset2.csv", index=False)
```
