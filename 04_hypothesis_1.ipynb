```python
import pandas as pd
df = pd.read_csv('dataset2.csv')
```

## 1. Hypothesis Test: ANOVA

**Test Method:** One-Way ANOVA (Analysis of Variance)

```python
from scipy.stats import f_oneway

groups = df.groupby("price_category")["revenue_usd"].apply(list)

f_stat, p = f_oneway(*groups)
print("{:.20e}".format(p))
print(f"p-value: {p:.5f}")
import numpy as np
print(np.format_float_scientific(p))
```

## Hypothesis Test Conclusion

**Decision Rule:** 
- If p-value (0.00) < 0.05 → **Reject H0** (statistically significant difference exists)

**Result:** 
- There is a significant difference in revenue across price category

## 2. Second Hypothesis Test: ANOVA (Units Sold)

**Test Method:** One-Way ANOVA (Analysis of Variance)
- Tests whether there are statistically significant differences in units sold between price categories

```python
from scipy.stats import f_oneway

groups = df.groupby("price_category")["units_sold"].apply(list)

f_stat, p = f_oneway(*groups)

print("p-value:", p)
```

## Second Test Conclusion

- p-value (0.21) ≥ 0.05 → **Fail to Reject H0** (no statistically significant difference)

**Result:** 
- No significant difference in units sold accross price category
