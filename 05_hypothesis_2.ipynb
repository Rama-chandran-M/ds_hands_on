```python
import pandas as pd 
df = pd.read_csv('dataset2.csv')
```

## 1. Hypothesis Test: Paired t-test (Discount Effect on Orders)

**Null Hypothesis (H0):** The mean number of orders with discount equals the mean number of orders without discount across models.
- H0: μ_discount = μ_no_discount

**Alternative Hypothesis (H1):** The mean number of orders differs between discounted and non-discounted purchases.
- H1: μ_discount ≠ μ_no_discount

**Test Method:** Paired t-test
- Compares order counts for the same models under two conditions: with and without discount
- Tests whether discount has a significant effect on the number of orders

```python
import pandas as pd
from scipy.stats import ttest_rel

df['discount_flag'] = df['discount_applied_status'].map({
    0: 'No Discount',
    1: 'Discount'
})

grouped = df.groupby(['model_name', 'discount_flag'])['order_id'].count().reset_index()

pivot_df = grouped.pivot(
    index='model_name',
    columns='discount_flag',
    values='order_id'
).fillna(0)
pivot_df = pivot_df[(pivot_df['No Discount'] > 0) & (pivot_df['Discount'] > 0)]

t_stat, p_value = ttest_rel(pivot_df['Discount'], pivot_df['No Discount'])

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("\nReject Null Hypothesis (H0)")
    print("Discount has a significant effect on number of orders")
else:
    print("\nFail to Reject Null Hypothesis (H0)")
    print("No significant effect of discount on orders")
```

## Interpretation
- Discount has a significant effect on number of orders placed

```python

avg_units_per_order = df.groupby(['model_name', 'discount_flag'])['units_sold'].mean().reset_index()

avg_units_pivot_df = avg_units_per_order.pivot(
    index='model_name',
    columns='discount_flag',
    values='units_sold'
).fillna(0)

avg_units_pivot_df = avg_units_pivot_df[(avg_units_pivot_df['No Discount'] > 0) & (avg_units_pivot_df['Discount'] > 0)]

t_stat_avg_units, p_value_avg_units = ttest_rel(avg_units_pivot_df['Discount'], avg_units_pivot_df['No Discount'])

print("Average Units Sold Per Order")
print("T-statistic:", t_stat_avg_units)
print("P-value:", p_value_avg_units)

if p_value_avg_units < alpha:
    print("\nReject Null Hypothesis (H0)")
    print("Discount has a significant effect on units per order")
else:
    print("\nFail to Reject Null Hypothesis (H0)")
    print("No significant effect of discount on units per order")
```

- 0.3 > 0.05
- Discount No significant effect on averge number of units sold in each order
