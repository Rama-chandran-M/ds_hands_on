```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('dataset2.csv')
```

```python
# Step 1: Create bins (5% interval)
df['discount_bin'] = pd.cut(
    df['discount_percent'],
    bins=range(0, int(df['discount_percent'].max()) + 5, 5)
)

# Step 2: Aggregate
discount_grouped = df.groupby('discount_bin')['units_sold'].sum()

# Step 3: Sort
discount_grouped = discount_grouped.sort_index()

# Step 4: Plot
import matplotlib.pyplot as plt

labels = [f"{int(i.left)}-{int(i.right)}" for i in discount_grouped.index]

plt.figure(figsize=(10,5))
plt.bar(labels, discount_grouped.values)

plt.xlabel('Discount Range (%)')
plt.ylabel('Total Units Sold')
plt.title('Sales by Discount Range')

plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
rating_grouped = df.groupby('customer_rating')['units_sold'].sum().sort_index()

plt.figure(figsize=(8,5))
plt.plot(rating_grouped.index, rating_grouped.values, marker='o')

plt.xlabel('Customer Rating')
plt.ylabel('Total Units Sold')
plt.title('Sales vs Customer Rating')

plt.grid()
plt.show()
```

```python
category_grouped = df.groupby('category')['units_sold'].sum()

plt.figure(figsize=(8,5))
plt.bar(category_grouped.index, category_grouped.values)

plt.xlabel('Category')
plt.ylabel('Total Units Sold')
plt.title('Sales by Category')

plt.xticks(rotation=45)
plt.show()
```

```python
channel_grouped = df.groupby('sales_channel')['units_sold'].sum()

plt.figure(figsize=(6,4))
plt.bar(channel_grouped.index, channel_grouped.values)

plt.xlabel('Sales Channel')
plt.ylabel('Total Units Sold')
plt.title('Sales by Channel')

plt.show()
```

```python
country_grouped = df.groupby('country')['revenue_usd'].sum()

plt.figure(figsize=(10,5))
plt.bar(country_grouped.index, country_grouped.values)

plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.title('Revenue by Country')

plt.xticks(rotation=45)
plt.show()
```

```python
income_grouped = df.groupby('customer_income_level')['units_sold'].sum()

plt.figure(figsize=(6,4))
plt.bar(income_grouped.index, income_grouped.values)

plt.xlabel('Income Level')
plt.ylabel('Total Units Sold')
plt.title('Sales by Income Level')

plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['month'] = df['order_date'].dt.month

month_grouped = df.groupby('month')['units_sold'].sum()

plt.figure(figsize=(8,5))
plt.plot(month_grouped.index, month_grouped.values, marker='o')

plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.title('Monthly Sales Trend')

plt.grid()
plt.show()
```
